```python
'''
import boto3
from pathlib import Path

# ---- Config ----
bucket = "housing-regression-data"   # replace with your bucket name
region = "eu-west-2"                 # your AWS region

# Set project root as parent of the notebooks folder
PROJECT_ROOT = Path.cwd().parent
local_data_dir = PROJECT_ROOT / "data" / "processed"
local_model_dir = PROJECT_ROOT / "models"

s3 = boto3.client("s3", region_name=region)

# ---- Helper function ----
def upload_file(local_path: Path, s3_key: str):
    if not local_path.exists():
        print(f"❌ File not found: {local_path}")
        return
    print(f"⬆️ Uploading {local_path} → s3://{bucket}/{s3_key}")
    s3.upload_file(str(local_path), bucket, s3_key)

# ---- Upload required datasets ----
upload_file(local_data_dir / "feature_engineered_holdout.csv", "processed/feature_engineered_holdout.csv")
upload_file(local_data_dir / "cleaning_holdout.csv", "processed/cleaning_holdout.csv")
upload_file(local_data_dir / "feature_engineered_train.csv", "processed/feature_engineered_train.csv")

# ---- Upload model ----
upload_file(local_model_dir / "xgb_best_model.pkl", "models/xgb_best_model.pkl") '''

import shutil
from pathlib import Path

# ---- Config (No AWS Required!) ----
# We create a fake local cloud storage folder inside your project
PROJECT_ROOT = Path.cwd().parent
MOCK_S3_BUCKET = PROJECT_ROOT / "mock_cloud_storage"

local_data_dir = PROJECT_ROOT / "data" / "processed"
local_model_dir = PROJECT_ROOT / "models"

# ---- Helper function using standard file copying ----
def upload_file_locally(local_path: Path, s3_key: str):
    if not local_path.exists():
        print(f"❌ File not found: {local_path}")
        return
    
    # Define where the file should go inside our fake cloud storage
    destination_path = MOCK_S3_BUCKET / s3_key
    
    # Create target folders if they don't exist yet
    destination_path.parent.mkdir(parents=True, exist_ok=True)
    
    print(f"📦 [LOCAL MOCK] Copying {local_path.name} → {destination_path}")
    shutil.copy(local_path, destination_path)

# ---- Upload required datasets ----
upload_file_locally(local_data_dir / "feature_engineered_holdout.csv", "processed/feature_engineered_holdout.csv")
upload_file_locally(local_data_dir / "cleaning_holdout.csv", "processed/cleaning_holdout.csv")
upload_file_locally(local_data_dir / "feature_engineered_train.csv", "processed/feature_engineered_train.csv")

# ---- Upload model ----
upload_file_locally(local_model_dir / "xgb_best_model.pkl", "models/xgb_best_model.pkl")
```
